```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File to Load (Remember to Change These)
file_to_load = os.path.join("Resources","purchase_data.csv")

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)
df
```

## Player Count
* Display the total number of players


```python
# Finding the number of players in the data frame
num_of_players_counted = len(df.SN.value_counts())
num_of_players_counted

# Cleaning up and putting it into a dataframe

player_count_df = pd.DataFrame({"Total Number of Players": [num_of_players_counted]})
player_count_df
```

## Purchasing Analysis (Total)
* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Need to find the:
#     Number of Unique items
#     Average price
#     Number of Purchases
#     Total Revenue
# print(df)
#     Number of Unique items
number_of_unique_items = df["Item Name"].nunique()
print(f"Number of Unique Items: {number_of_unique_items}")

#     Average price
mean_price_of_items = round(df.Price.mean(),2)
print(f"Mean/average price: ${mean_price_of_items}")

#     Number of Purchases
number_of_purcahses = df["Purchase ID"].count()
print(f"Number of purchases: {number_of_purcahses}")

#     Total Revenue
total_revenue_of_items = round(df.Price.sum(),2)
print(f"The Total Revenue: ${total_revenue_of_items}")

# Making it presentable
purchasing_analysis_df = pd.DataFrame({"Number of Unique Items":[number_of_unique_items],
                                       "Mean Price":[mean_price_of_items],
                                       "Number Of Purchases":[number_of_purcahses],
                                       "Total Revenue":[total_revenue_of_items],
                                      })
purchasing_analysis_df.style.format({"Mean Price":"${:,.2f}",
                                     "Total Revenue":"${:,.2f}"})
# And there we go
```

## Gender Demographics
* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed



```python
# Gender Demographics

# Grouping the genders
genders_df = df.groupby("Gender")

#sanity check
genders_df.head()

# finding the number of genders with respect to their User ID/SN
genders_count = genders_df.nunique()["SN"]
# Sanity check for genders count
print(genders_count)
# Other Sanity Checks
m = "Male"
f = "Female"
o = "Other / Non-Disclosed"
print(f"Number of Male Players: {genders_count[m]}")
print(f"Number of Male Players: {genders_count[f]}")
print(f"Number of not male and not female: {genders_count[o]}")
print(f"Total number of players: {num_of_players_counted}")
# Now I know how it Works, lets work out the percentages
# Also doing a Sanity check

#     Percentage and Count of Male Players
male_percentage = (genders_count[m]/num_of_players_counted)*100
print(f"Male percentage: {male_percentage}")
#     Percentage and Count of Female Players
female_percentage = (genders_count[f]/num_of_players_counted)*100
print(f"Female percentage: {female_percentage}")
#     Percentage and Count of Other / Non-Disclosed
other_percentage = (genders_count[o]/num_of_players_counted)*100
print(f"Other percentage: {other_percentage}")

gender_analysis_df = pd.DataFrame({"Total Count":[genders_count[m],genders_count[f],genders_count[o]],
                                   "Player Percentage":[male_percentage,female_percentage,other_percentage]},
                                  index=[m,f,o])

gender_analysis_df.index.name = None

gender_analysis_df.style.format({'Player Percentage':'{:,.2f}%'})

# And there we go
```

## Age Demographics
* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table

```python
#     Establish bins for ages
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 9999]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]

#     Categorize the existing players using the age bins. Hint: use pd.cut()
df["Age Range"] = pd.cut(df["Age"], bins, labels=labels)
df
#     Calculate the numbers and percentages by age group

# Counting
age_range_grouped = df.groupby("Age Range")
age_range_grouped_count = age_range_grouped.SN.nunique()
# Sanity Check
# print(age_range_grouped_count)

# Finding the Percentage
age_range_grouped_percent = (age_range_grouped_count/num_of_players_counted)*100
# print(age_range_grouped_percent)
#     Create a summary data frame to hold the results
age_demo_df = pd.DataFrame({"Total Count":age_range_grouped_count,
                            "Player Percentage":age_range_grouped_percent})
age_demo_df
#     Optional: round the percentage column to two decimal points
age_demo_df["Player Percentage"] = age_demo_df["Player Percentage"].map("{:.2f}%".format)
#     Display Age Demographics Table
age_demo_df
```

## Top Spenders

* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order

```python
# Top Spenders
groupby_spender = df.groupby('SN')

#     Run basic calculations to obtain the results in the table below

# Finding the Purchase Count per player/spender
groupby_spender_count = groupby_spender["Purchase ID"].count()
print(f"{groupby_spender_count}\n")

# Finding the Average purchase price per player/spender
groupby_spender_mean = groupby_spender.Price.mean()
print(f"{groupby_spender_mean}\n")

# Finding the Total Purchase Value per player/spender
groupby_spender_sum = groupby_spender.Price.sum()
print(f"{groupby_spender_sum}\n")

#     Create a summary data frame to hold the results
spending_purchasing_analysis = pd.DataFrame({"Purchase Count":groupby_spender_count,
                                     "Average Purchase Price":groupby_spender_mean,
                                     "Total Purchase Value":groupby_spender_sum})

#     Optional: give the displayed data cleaner formatting
# Making the data a bit more pretty
spending_purchasing_analysis["Average Purchase Price"] = spending_purchasing_analysis["Average Purchase Price"].map("${:.2f}".format)
spending_purchasing_analysis["Total Purchase Value"] = spending_purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format)

#     Sort the total purchase value column in descending order
decending_spending_purchasing_analysis = spending_purchasing_analysis.sort_values(["Purchase Count"], ascending=False)
#     Display a preview of the summary data frame

decending_spending_purchasing_analysis.head()
```

## Most Popular Items
* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Most Popular Items

#     Retrieve the Item ID, Item Name, and Item Price columns
items = df[["Item ID","Item Name","Price"]]

#     Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
groupby_items = items.groupby(["Item ID","Item Name"])

# Item Count
groupby_item_count = groupby_items.Price.count()
# Sanity Check
print(groupby_item_count)

# Total Item sum in USD
groupby_item_sum = (groupby_items.Price.sum())
# Sanity Check
print(groupby_item_sum)

# Item Prices
item_prices = (groupby_item_sum/groupby_item_count)
# Sanity Check
print(item_prices)

item_analysis = pd.DataFrame({"Purchase Count": groupby_item_count,
                              "Item Price":item_prices,
                              "Total Purchase Value":groupby_item_sum})

#     Optional: give the displayed data cleaner formatting
# Making the data a bit more pretty
item_analysis["Item Price"] = item_analysis["Item Price"].map("${:.2f}".format)
item_analysis["Total Purchase Value"] = item_analysis["Total Purchase Value"].map("${:,.2f}".format)

#     Sort the total purchase value column in descending order
decending_item_analysis = item_analysis.sort_values(["Purchase Count"], ascending=False)

#     Display a preview of the summary data frame
decending_item_analysis.head()
```
